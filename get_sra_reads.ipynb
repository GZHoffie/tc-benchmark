```python
from pysradb import SRAweb

# We collect reads from the following study
"""
Study: Genome sequencing and hybrid assembly of microbial isolates derived from the human gut.
PRJNA885826 • SRP438478
Abstract: We generated a collection of complete or near complete genomes of microbial isolates 
derived from the human gut using a combination of Illumina paired end reads and Oxford nanopore long reads.
"""

db = SRAweb()
study_df = db.sra_metadata("SRP438478")
```

```python
study_df = study_df[study_df["instrument_model_desc"] == "OXFORD_NANOPORE"]
study_df.head()
```

```python
study_df.columns
```

```python
db.download(df=study_df, out_dir='/pysradb_downloads')()
```
