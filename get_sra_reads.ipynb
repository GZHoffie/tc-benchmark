```python
from pysradb import SRAweb

db = SRAweb()
```

```python
import pandas as pd
from pysradb.search import EnaSearch, SraSearch
import numpy as np
```

```python
import pandas as pd

gtdb_metadata = pd.read_csv('./bac120_metadata.tsv',sep='\t')
```

```python
species_df = gtdb_metadata[["ncbi_taxid", "ncbi_organism_name"]]
species_df = species_df.groupby("ncbi_taxid").sample(1)
species_df
```

```python

species_df.iloc[1]['ncbi_taxid']
```

```python
instance = SraSearch(2, 10000, query="R10.4.1", source="genomic", strategy="wgs", platform="oxford nanopore")
instance.search()
res_df = instance.get_df()
```

```python
res_df
```

```python
sample_df = res_df.groupby("sample_taxon_id").sample(1)
sample_df
```

```python
from tqdm import tqdm

run_sample_info_df = None

for i in tqdm(range(len(sample_df))):
    study_df = db.sra_metadata(sample_df.iloc[i]["study_accession"], detailed=True)
    study_df_sample = study_df.sample(1)

    if run_sample_info_df is None:
        run_sample_info_df = study_df_sample
    else:
        run_sample_info_df = pd.concat([run_sample_info_df, study_df_sample])
```

```python
run_sample_info_df = run_sample_info_df[run_sample_info_df["instrument_model_desc"] == "OXFORD_NANOPORE"]
run_sample_info_df.to_csv("nanopore_10.4_data_samples.csv")
```

```python
import subprocess
import os

num_samples = 1000

os.chdir("./data")
for i in range(len(run_sample_info_df)):
    subprocess.run(["fastq-dump", str(run_sample_info_df.loc[i]["run_accession"]), "-X", str(num_samples)])
```
